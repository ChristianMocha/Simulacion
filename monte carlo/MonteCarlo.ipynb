```python
import random
import numpy as np
import matplotlib.pyplot as plt
def monteCarlo(total, r):
    hit = 0
    x_hit = []
    y_hit = []
    x_miss = []
    y_miss = []
    n = 0
    
    for i in range(total):
        # inisiasi x dan y secara random
        rand_x = random.uniform(0,r)
        rand_y = random.uniform(0,r)
        
        # hitung x dan y
        tmp = rand_x**2 + rand_y**2
        
        # cek kondisi hit
        if tmp <= r**2:
            x_hit.append(rand_x) 
            y_hit.append(rand_y) 
            hit += 1 
        else:
            x_miss.append(rand_x) 
            y_miss.append(rand_y) 
        n += 1
        
    # hitung nilai pi
    pi = 4 * (hit / total)
        
    return pi, x_hit, y_hit, x_miss, y_miss
        

pi, xh_arr, yh_arr, xm_arr, ym_arr = monteCarlo(5000, 1)
print("Pi: ", pi)

plt.title("N = 5000")
plt.scatter(xh_arr, yh_arr, c='y', s=2.5)
plt.scatter(xm_arr, ym_arr, c='b', s=2.5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
